### Model Building

```python
# Finding missing values in dataset
telco_data.isnull().sum()
```

```python
# Missing value imputation
#telco_data.dropna()      -- the missing values are very small so we can either drop that rows because it doesn't effect predictions
# But here,we are imputing values TotalCharges with MonthlyCharges of that customers
telco_data['TotalCharges'].fillna(telco_data['MonthlyCharges'],inplace=True)
```

```python
telco_data.head()
```

```python
# FOr better performaance of model we are converting churn categorical data no,yes to numeric 0,1
telco_data['Churn'].replace({'Yes':1,'No':0},inplace=True)
```

```python
telco_data.info()
```

```python
# Feature Extraction: By using Contract we are defining new column Is_Premium for customers
telco_data['Is_Premium']=telco_data["Contract"].map({'Month-to-month':'No','One year':'Yes','Two year':'Yes'})
```

```python
telco_data
```

### Encoding

```python
# By using dummy values convert categorical into numeric variables
df_dummies=pd.get_dummies(telco_data, drop_first=True)
df_dummies.head()
```

```python
# Checking if there is any outliers are present in dataset
sns.boxplot(telco_data)
```

By above boxplot we conclude there are not much outliers are present

```python
X=df_dummies.drop(columns=['Churn'])
y=df_dummies['Churn'].values
```

```python
# Spliting data into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=50)
```

```python
# Data scaling
from sklearn.preprocessing import MinMaxScaler
features=X_train.columns.values
scaler=MinMaxScaler(feature_range=(0,1))
scaler.fit(X_train)
X_train=pd.DataFrame(scaler.transform(X_train))
X_train.columns=features
```

```python
X_test=pd.DataFrame(scaler.transform(X_test))
X_test.columns=features
```

```python
# Running Logistic Regression Model
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model_lr=model.fit(X_train,y_train)
```

```python
y_pred_lr=model.predict(X_test)
```

```python
from sklearn import metrics
print('Classification Report:',metrics.classification_report(y_test,y_pred_lr))
print("Confusion Matrix:",metrics.confusion_matrix(y_test,y_pred_lr))
```

```python
# Running Random Forest Model
from sklearn.ensemble import RandomForestClassifier
model_rf=RandomForestClassifier(n_estimators=1000,oob_score=True,n_jobs=-1,random_state=50,max_features=20,max_leaf_nodes=30)
model_rf.fit(X_train,y_train)
y_pred_rf=model_rf.predict(X_test)
print('Classification Report:',metrics.classification_report(y_test,y_pred_rf))
print("Confusion Matrix:",metrics.confusion_matrix(y_test,y_pred_rf))
```

```python
#Support Vector Machine
from sklearn.svm import SVC
model.svm=SVC(kernel='linear')
model.svm.fit(X_train,y_train)
y_pred_svm=model.svm.predict(X_test)
print('Classification Report:',metrics.classification_report(y_test,y_pred_svm))
print("Confusion Matrix:",metrics.confusion_matrix(y_test,y_pred_svm))
```

```python
# Ada Boost
from sklearn.ensemble import AdaBoostClassifier
model_adb=AdaBoostClassifier()
model_adb.fit(X_train,y_train)
y_pred_adb=model_adb.predict(X_test)
print('Classification Report:',metrics.classification_report(y_test,y_pred_adb))
print("Confusion Matrix:",metrics.confusion_matrix(y_test,y_pred_adb))
```
